```python
%config IPCompleter.greedy = True
from pyspark.sql import SparkSession

spark = (SparkSession.builder
    .appName("test")
    .config("spark.master", "local[10]")
    .config("spark.executor.memory", "8g")
    .config("spark.driver.memory", "8g")
    .config("spark.sql.shuffle.partitions", "40") # для dataframe
    .config("spark.default.parallelism", "40") # для rdd
    .config("spark.sql.adaptive.enabled", "true") # aqe
    .config("spark.sql.adaptive.coalescePartitions.enabled", "true") # aqe для coalesce партиций
    .config("spark.sql.adaptive.advisoryPartitionSizeInBytes", "128MB") # целевое значение
    .config("spark.sql.autoBroadcastJoinThreshold", "-1")
    .getOrCreate()
)
```

```python
df = spark.sql('select 1 as number')
```

```python
# создать датафрейм
data = [("Alice", 25), ("Bob", 30), ("Cathy", 35)]
df = spark.createDataFrame(data, ["name", "age"])
df.show()
```
